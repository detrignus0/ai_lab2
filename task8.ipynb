```python
import numpy as np

matrix = np.random.randint(1, 11, size=(3, 3))
print("1. Исходная матрица (3x3):\n", matrix)

det = np.linalg.det(matrix)
print("\n2. Определитель матрицы:", det)

if np.isclose(det, 0):
    print("\n3. Матрица вырождена, обратной матрицы не существует.")
else:
    inv_matrix = np.linalg.inv(matrix)
    print("\n3. Обратная матрица:\n", inv_matrix)
```
