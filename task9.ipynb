```python
import numpy as np

matrix = np.random.randint(1, 51, size=(4, 4))
print("1. Исходная матрица (4x4):\n", matrix)

transposed_matrix = matrix.T
print("\n2. Транспонированная матрица:\n", transposed_matrix)

matrix_trace = np.trace(matrix)
print("\n3. След матрицы (сумма элементов главной диагонали):", matrix_trace)
```
