```python
import numpy as np

A = np.array([
    [2, 3, -1],
    [4, -1, 2],
    [-3, 5, 4]
])

B = np.array([5, 6, -2])

print("1. Матрица коэффициентов A:\n", A)
print("\n2. Вектор правой части B:", B)

X = np.linalg.solve(A, B)

print("\n3. Решение системы (вектор X):")
print("   x =", X[0])
print("   y =", X[1])
print("   z =", X[2])
```
