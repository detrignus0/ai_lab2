```python
import numpy as np

matrix_A = np.random.randint(1, 21, size=(3, 3))
matrix_B = np.random.randint(1, 21, size=(3, 3))

print("1. Матрица A (3x3):\n", matrix_A)
print("\n2. Матрица B (3x3):\n", matrix_B)

matrix_sum = matrix_A + matrix_B
matrix_diff = matrix_A - matrix_B
matrix_elementwise_prod = matrix_A * matrix_B

print("\n3. Результаты операций:")
print("   Сумма матриц:\n", matrix_sum)
print("\n   Разность матриц:\n", matrix_diff)
print("\n   Поэлементное произведение:\n", matrix_elementwise_prod)
```
