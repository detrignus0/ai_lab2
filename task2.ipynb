```python
import numpy as np

array = np.random.randint(1, 101, size=20)
print("1. Исходный случайный массив:\n", array)

mask = (array % 5 == 0)

divisible_by_5 = array[mask]
print("\n2. Элементы, делящиеся на 5 без остатка:\n", divisible_by_5)

array[mask] = -1
print("\n3. Обновленный массив (числа, кратные 5, заменены на -1):\n", array)
```
